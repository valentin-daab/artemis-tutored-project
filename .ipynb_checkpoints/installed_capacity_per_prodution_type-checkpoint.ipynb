```python
eurostat = pd.read_excel('DATA/EUROSTAT/installed_capacity_per_production_type.xlsx',      
             sheet_name='Sheet 1')
eurostat
```

```python
years = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007',
         '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015',
        '2016', '2017', '2018', '2019', '2020']



eurostat = eurostat.set_index(['GEO (Labels)', 'SIEC (Labels)'])
eurostat = pd.melt(eurostat, value_vars = years, ignore_index=False)
eurostat = eurostat.reset_index('SIEC (Labels)')
eurostat = eurostat.set_index('variable', append = True)
eurostat = eurostat.pivot(columns = 'SIEC (Labels)')
eurostat
```
