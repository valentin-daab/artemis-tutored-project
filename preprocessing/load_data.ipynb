# Actual generation per production type

```python
df = pd.DataFrame()

start = pd.Timestamp('20150101', tz='Europe/Brussels')
end = pd.Timestamp('20200101', tz='Europe/Brussels')

country_code = ['AT', 'FR', 'DE', 'UK', 'LU', 'RO']

for count, value in enumerate(country_code) :
    temp_df = client.query_generation(value, start=start,end=end, psr_type=None)
    temp_df.insert(0, 'Country', value)
    df = pd.concat([df, temp_df])
    print(value)
    temp_df = pd.DataFrame()
```
