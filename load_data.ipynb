# Installed Capacity per Production Type

```python
import pandas as pd

from datetime import timedelta
from entsoe import EntsoePandasClient
from IPython.core.display import HTML
```

```python
client = EntsoePandasClient(api_key="e72b5f2d-2b4a-4eed-ba56-f05452380ee6")

start = pd.Timestamp('20000101', tz='Europe/Brussels')
end = pd.Timestamp('20200102', tz='Europe/Brussels')

country_code = ['AL', 'AT', 'BE', 'BA', 'BG', 'HR', 'CY', 'CZ', 'DK',
               'EE', 'FI', 'FR', 'DE', 'GR', 'HU', 'IE', 'IT',
               'LV', 'LT', 'LU', 'ME', 'NL', 'MK', 'NO', 'PL',
               'PT', 'RO', 'RS', 'SK', 'SI', 'ES', 'SE', 'CH', 'UA',
               'UK']

#out : GE, XK
```

```python
df = pd.DataFrame()


for count, value in enumerate(country_code) :
    temp_df = client.query_installed_generation_capacity(value, start=start,end=end, psr_type=None)
    temp_df.insert(0, 'Country', value)
    df = pd.concat([df, temp_df])
    print(value)
    temp_df = pd.DataFrame()
```

```python
df = df.reset_index()
df = df.rename(columns = {'index' : 'Date'})
df['Date'] = pd.to_datetime(df['Date'], utc = True)
df['Date'] = df['Date'].apply(lambda x: x + timedelta(days=1))
df['Date'] = df['Date'].dt.tz_localize(None)
df['Date'] = pd.to_datetime(df['Date']).dt.date
df.to_excel('installed_capacity_per_production_type.xlsx')
```
